```python
data_dir = './data/csv'

data_entry_file = os.path.join(data_dir, 'Data_Entry_2017_v2020.csv')

data_entry_df = pd.read_csv(data_entry_file)
```

```python
sample_size = data_entry_df_copy.shape[0]

pneumonia_cases = data_entry_df_copy.query('pneumonia == 1')
nonpneumonia_cases = data_entry_df_copy.query('pneumonia == 0')

num_pneumonia_cases = pneumonia_cases.shape[0]
num_nonpneumonia_cases = nonpneumonia_cases.shape[0]

pct_pneumonia_cases = num_pneumonia_cases / sample_size
pct_nonpneumonia_cases = num_nonpneumonia_cases / sample_size
```

```python
print('Total number of records: {}'.format(sample_size))

print('Total number of Pneumonia cases: {}'.format(num_pneumonia_cases))
print('Total number of non Pneumonia cases: {}'.format(num_nonpneumonia_cases))

print('Percentage of Pneumonia: {:0.2f}'.format(pct_pneumonia_cases))
print('Percentage of non Pneumonia: {:0.2f}'.format(pct_nonpneumonia_cases))
```

```python
training_size = int(np.rint(sample_size * 0.8))
testing_size = sample_size - training_size

num_pneumonia_cases_in_training = int(np.rint(training_size * pct_pneumonia_cases))
num_pneumonia_cases_in_testing = num_pneumonia_cases - num_pneumonia_cases_in_training

num_nonpneumonia_cases_in_training = training_size - num_pneumonia_cases_in_training
num_nonpneumonia_cases_in_testing = testing_size - num_pneumonia_cases_in_testing
```

```python
print('Total number of training cases: {}'.format(training_size))
print('Total number of Pneumonia cases in training: {}'.format(num_pneumonia_cases_in_training))
print('Total number of non Pneumonia cases in training: {}'.format(num_nonpneumonia_cases_in_training))

print('Total number of testing cases: {}'.format(testing_size))
print('Total number of Pneumonia cases in testing: {}'.format(num_pneumonia_cases_in_testing))
print('Total number of non Pneumonia cases in testing: {}'.format(num_nonpneumonia_cases_in_testing))
```

```python
pneumonia_cases_in_training = pneumonia_cases.sample(num_pneumonia_cases_in_training)
nonpneumonia_cases_in_training = nonpneumonia_cases.sample(num_nonpneumonia_cases_in_training)

training_df = pd.concat([pneumonia_cases_in_training, nonpneumonia_cases_in_training]).reset_index()

pneumonia_cases_in_training.shape[0], nonpneumonia_cases_in_training.shape[0]
```

```python
pneumonia_cases_in_testing = \
    pneumonia_cases[~pneumonia_cases.image_index.isin(pneumonia_cases_in_training['image_index'])]
nonpneumonia_cases_in_testing = \
    nonpneumonia_cases[~nonpneumonia_cases.image_index.isin(nonpneumonia_cases_in_training['image_index'])]

testing_df = pd.concat([pneumonia_cases_in_testing, nonpneumonia_cases_in_testing]).reset_index()

pneumonia_cases_in_testing.shape[0], nonpneumonia_cases_in_testing.shape[0]
```

```python
training_filepath = os.path.join(data_dir, "training_data.csv")

training_df.to_csv (training_filepath, index = True, header = True)
```

```python
testing_filepath = os.path.join(data_dir, "testing_data.csv")

testing_df.to_csv (testing_filepath, index = True, header = True)
```
