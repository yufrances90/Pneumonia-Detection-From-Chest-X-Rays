```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_dir = '../input/nih-data-generator'

test_file = os.path.join(data_dir, 'testing_data.csv')
train_file = os.path.join(data_dir, 'train_data.csv')
valid_file = os.path.join(data_dir, 'validation_data.csv')

train_df = pd.read_csv(train_file, index_col=0)
valid_df = pd.read_csv(valid_file, index_col=0)
test_df = pd.read_csv(test_file, index_col=0)
```

```python
current_palette = sns.color_palette("colorblind")

sns.set_palette(current_palette)

base_color=current_palette[0]
```

```python
fig, axs = plt.subplots(1, 3, figsize=(21, 8))

sns.countplot(train_df['pneumonia'], color=base_color, ax=axs[0])
axs[0].set_title('Training')

sns.countplot(valid_df['pneumonia'], color=base_color, ax=axs[1])
axs[1].set_title('Validation')

sns.countplot(test_df['pneumonia'], color=base_color, ax=axs[2])
axs[2].set_title('Testing')

plt.suptitle('Distribution of Pneumonia-positive and Pneumonia-negative cases')

plt.show()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(21, 8), sharey=True)

bins = np.arange(0, 95+10, 10)

axs[0].hist(train_df['patient_age'], bins=bins)
axs[0].set_title('Training')

axs[1].hist(valid_df['patient_age'], bins=bins)
axs[1].set_title('Validation')

axs[2].hist(test_df['patient_age'], bins=bins)
axs[2].set_title('Testing')

plt.xticks(bins)

plt.suptitle('Distribution of Patient Age')

plt.show()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(21, 8))

bins = np.arange(0, 95+10, 10)

gender_classes = ['F', 'M']

sns.countplot(data=train_df, x='patient_gender', color=base_color, ax=axs[0], order=gender_classes)
axs[0].set_title('Training')

sns.countplot(data=valid_df, x='patient_gender', color=base_color, ax=axs[1], order=gender_classes)
axs[1].set_title('Validation')

sns.countplot(data=test_df, x='patient_gender', color=base_color, ax=axs[2], order=gender_classes)
axs[2].set_title('Testing')

plt.suptitle('Distribution of Patient Age')

plt.show()
```
