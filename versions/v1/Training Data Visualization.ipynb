```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_dir = './archived/data/csv'

model_6_dir = os.path.join(data_dir, 'model_6')
model_7_dir = os.path.join(data_dir, 'model_7')

train_data_file = 'train_data.csv'
test_data_file = 'test_data.csv'
```

```python
current_palette = sns.color_palette("colorblind")

sns.set_palette(current_palette)

base_color=current_palette[0]
```

---
---

# Model 2

```python
train_file = os.path.join(model_6_dir, train_data_file)
test_file = os.path.join(model_6_dir, test_data_file)

train_df = pd.read_csv(train_file, index_col=0)
test_df = pd.read_csv(test_file, index_col=0)
```

```python
fig, axs = plt.subplots(2, 1, figsize=(10, 10), sharex=True)

train_finding_labels_sum = train_df['finding_labels'].str.get_dummies(sep='|').sum()
test_finding_labels_sum = test_df['finding_labels'].str.get_dummies(sep='|').sum()

train_y_pos = np.arange(len(train_finding_labels_sum))
test_y_pos = np.arange(len(test_finding_labels_sum))

axs[0].barh(train_y_pos, train_finding_labels_sum)
axs[0].set_yticks(train_y_pos)
axs[0].set_yticklabels(labels=train_finding_labels_sum.keys())

axs[1].barh(test_y_pos, test_finding_labels_sum)
axs[1].set_yticks(test_y_pos)
axs[1].set_yticklabels(labels=test_finding_labels_sum.keys())

plt.tight_layout(pad=1)

plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(16, 8), sharey=True)

bins = np.arange(0, 95+10, 10)

axs[0].hist(train_df['patient_age'], bins=bins)
axs[1].hist(test_df['patient_age'], bins=bins)

plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(16, 8), sharey=True)

bins = np.arange(0, 63340+5000, 5000)

sns.countplot(data=train_df, x='patient_gender', color=base_color, ax=axs[0])
sns.countplot(data=test_df, x='patient_gender', color=base_color, ax=axs[1])

plt.show()
```

```python
train_finding_labels_counts = train_df['finding_labels'].value_counts()
test_finding_labels_counts = test_df['finding_labels'].value_counts()

train_finding_labels_counts = train_finding_labels_counts.sort_values(ascending=False)
test_finding_labels_counts = test_finding_labels_counts.sort_values(ascending=False)

train_first_10_finding_labels = train_finding_labels_counts[:10]
test_first_10_finding_labels = test_finding_labels_counts[:10]

fig, axs = plt.subplots(1, 2, figsize=(20, 10))

sns.barplot(
    x=train_first_10_finding_labels,
    y=train_first_10_finding_labels.keys(), 
    color=base_color,
    order=train_first_10_finding_labels.index,
    ax=axs[0]
)

sns.barplot(
    x=test_first_10_finding_labels,
    y=test_first_10_finding_labels.keys(), 
    color=base_color,
    order=test_first_10_finding_labels.index,
    ax=axs[1]
)

plt.tight_layout(pad=1)

plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(16, 8), sharey=True)

bins = np.arange(0, 67310+5000, 5000)

sns.countplot(data=train_df, x='view_position', color=base_color, ax=axs[0])
sns.countplot(data=train_df, x='view_position', color=base_color, ax=axs[1])

plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(16, 8), sharey=True)

sns.countplot(data=train_df, x='pneumonia', color=base_color, ax=axs[0])
sns.countplot(data=test_df, x='pneumonia', color=base_color, ax=axs[1])

fig.suptitle('Model 2: Distribution of Pneumonia-positive and Pneumonia-negative images')

plt.show()
```
